Generación de muestras de una variable aleatoria normal usando numpy.

```python
pip install statsmodels
```

```python
import numpy as np

mu = 3
sigma = 0.5
n=1000
vals = np.random.lognormal(mean=mu, sigma=sigma, size=n)
print(vals)
```

Cálculo de estadísticas usando pandas.

```python
import pandas as pd

df = pd.DataFrame(vals)
print(df.describe())
```

Generación de histograma usando matplotlib.

```python
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(x=vals, bins=30)
plt.title('Histograma de tiempos de servicio')
plt.xlabel('Tiempos de servicio')
plt.ylabel('Frecuencia')
plt.show()

print(bins)
print(count)
```

Densidad normal estándar.

```python
from scipy.stats import lognorm
import numpy as np
import matplotlib.pyplot as plt

# Normal (mu=0, sigma=1)
x= np.arange(0,4,0.001)
plt.plot(x, lognorm.pdf(x, 1, np.exp(0)))
plt.title('Función de densidad lognormal estándar')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

Densidad normal con media 3 y desviación estándar 1.

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

# Normal (mu=3, sigma=1)
x= np.arange(1,6,0.01)
plt.plot(x, lognorm.pdf(x, 1, np.exp(3)))
plt.title('Función de Densidad Log-Normal (mu=3, sigma=1)')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

Densidad normal con media 3 y desviación estándar 0.5.

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

# Normal (mu=3, sigma=1)
x= np.arange(1,6,0.01)
plt.plot(x, lognorm.pdf(x, 0.5, np.exp(1)))
plt.title('Función de Densidad Log-Normal (mu=1, sigma=0.5)')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

QQ plot usando statsmodels. 

```python
import statsmodels.api as sm
import scipy.stats as stats
fig = sm.qqplot(data=vals, dist=stats.lognorm, distargs=(1,), loc=np.exp(3), scale=1, line="45")
plt.show()
```

QQ plot contra normal con media 3 y desviación estándar 1.

```python
import statsmodels.api as sm
import scipy.stats as stats
fig = sm.qqplot(data=vals, dist=stats.lognorm, distargs=(1,), loc=np.exp(3), scale=1, line="45")
plt.show()
```

QQ plot contra normal con media 0 y desviación estándar 1.

```python
import statsmodels.api as sm
import scipy.stats as stats
fig = sm.qqplot(data = vals, dist=stats.norm, loc=0, scale=1, line="45")
plt.show()
```
